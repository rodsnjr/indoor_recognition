```python
import utils as ut
import models as md
from keras.callbacks import ModelCheckpoint, LearningRateScheduler, TensorBoard, EarlyStopping
from keras.models import Sequential, Model 
from keras.layers import Dropout, Flatten, Dense, GlobalAveragePooling2D
from keras_tqdm import TQDMCallback

train_x, train_labels = ut.read_file('%s_mobilenet' % ut.Directory.train.name)
train_y, _ = ut.get_data_labels(train_labels)

print(train_x.shape)

validation_x, validation_labels = ut.read_file('%s_mobilenet' % ut.Directory.validation.name)
validation_y, _ = ut.get_data_labels(validation_labels)

checkpoint = ModelCheckpoint("mobilenet.h5", monitor='val_acc', verbose=1, save_best_only=True, save_weights_only=False, mode='auto', period=1)
early = EarlyStopping(monitor='val_acc', min_delta=0, patience=10, verbose=1, mode='auto')
progress = TQDMCallback(leave_outer=False)

epochs = 200
batch_size=32

clf_model = Sequential()
clf_model.add(GlobalAveragePooling2D(input_shape=train_x.shape[1:]))
clf_model.add(Dense(1024, activation="relu"))
clf_model.add(Dense(1024, activation="relu"))
clf_model.add(Dense(5, activation="softmax"))
clf_model.compile(loss = "categorical_crossentropy", 
              optimizer = "adam", 
              metrics=["categorical_accuracy", 'mae'])

clf_model.fit(train_x, train_y,
         batch_size=batch_size,
        epochs=epochs, verbose=0,
       validation_data=(validation_x, validation_y),
    callbacks = [checkpoint, early, progress])

clf_model.save('mobilenet.h5')
```
