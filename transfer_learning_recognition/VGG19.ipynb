```python
import utils as ut
import models as md
from keras.models import Sequential, Model 
from keras.layers import Dropout, Flatten, Dense, GlobalAveragePooling2D, GlobalMaxPooling2D
from keras.callbacks import ModelCheckpoint, LearningRateScheduler, TensorBoard, EarlyStopping
from keras_tqdm import TQDMCallback

train_x, train_labels = ut.read_file('%s_vgg19' % ut.Directory.train.name)
train_y, _ = ut.get_data_labels(train_labels)

validation_x, validation_labels = ut.read_file('%s_vgg19' % ut.Directory.validation.name)
validation_y, _ = ut.get_data_labels(validation_labels)

checkpoint = ModelCheckpoint("vgg16_1.h5", monitor='val_acc', verbose=1, save_best_only=True, save_weights_only=False, mode='auto', period=1)
early = EarlyStopping(monitor='val_acc', min_delta=0, patience=10, verbose=1, mode='auto')
progress = TQDMCallback(leave_outer=False)

epochs = 100
batch_size=64

clf_model = Sequential()
clf_model.add(GlobalMaxPooling2D(input_shape=train_x.shape[1:]))
# clf_model.add(Flatten(input_shape=train_x.shape[1:]))
clf_model.add(Dense(2048, activation="relu"))
clf_model.add(Dense(2048, activation="relu"))
clf_model.add(Dense(5, activation="softmax"))
clf_model.compile(loss = "categorical_crossentropy", 
              optimizer = "adam", 
              metrics=["accuracy"])

clf_model.fit(train_x, train_y,
          batch_size=batch_size,
          epochs=epochs, verbose=0,
          validation_data=(validation_x, validation_y),
        callbacks = [checkpoint, early, progress])
```

```python
import utils as ut
import models as md
from keras.models import load_model

test_x, test_labels = ut.read_file('%s_vgg19' % ut.Directory.test.name)
test_y, _ = ut.get_data_labels(test_labels)


model = load_model('vgg16_1.h5')
model.evaluate(test_x, test_y, batch_size=64)
```
